**SHORT DESCRIPTION ABOUT CHURN MODELLING DATASET**


1. *CustomerID*: A unique identifier for each customer.

2. *Surname*: The customer's last name.

3. *CreditScore*: A numeric score representing the creditworthiness of the customer.

4. *Geography*: The country or region where the customer resides.

5. *Gender*: The gender of the customer (e.g., Male, Female).

6. *Age*: The age of the customer.

7. *Tenure*: The number of years the customer has been with the company.

8. *Balance*: The amount of money the customer currently has in their account(s).

9. *NumOfProducts*: The number of products the customer has purchased or is subscribed to.

10. *HasCrCard*: A binary indicator (Yes/No or 1/0) indicating whether the customer has a credit card.

11. *IsActiveMember*: A binary indicator showing if the customer is an active member (Yes/No or 1/0).

12. *EstimatedSalary*: The estimated annual salary of the customer.

13. *Exited*: The target variable, indicating whether the customer has left the company (Yes/No or 1/0).
